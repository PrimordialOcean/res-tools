```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_data(filename):
    df = pd.read_csv(filename, encoding='utf-8', sep=',')
    tempc_olv = df['TempC_olv']
    tempc_plg = df['TempC_plg']
    h2o_olv = df['H2O_olv']
    h2o_plg = df['H2O_plg']
    return tempc_olv, tempc_plg, h2o_olv, h2o_plg


def plot(tempc_olv, tempc_plg, h2o_olv, h2o_plg):
    plt.style.use('default')
    plt.rcParams["font.size"] = 16
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'
    fig = plt.figure( figsize=(4.0, 4.0) )
    ax = fig.add_subplot(111) 
    
    ax.plot(tempc_plg, h2o_plg, '-', label='Plg-liq')
    ax.plot(tempc_olv, h2o_olv, '--', label='Olv-liq')
    
    ax.set_xlabel('Temperature [Celsius]')
    ax.set_ylabel('H$_2$O [wt.%]')
    
    plt.legend()
    plt.show()
    

def main():    
    filename = 'data.csv'
    tempc_olv, tempc_plg, h2o_olv, h2o_plg = read_data(filename)
    plot(tempc_olv, tempc_plg, h2o_olv, h2o_plg)
    
    
if __name__ == "__main__":
    main()
```
