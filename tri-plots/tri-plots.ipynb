```python
# -*- coding: utf-8 -*-
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sheet = [ 'px(gm)', 'opx', 'cpx' ] # シートに分けて入力すると自動で色を変える
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect('equal')

# 枠線を消す
plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.tick_params(bottom=False, left=False, right=False, top=False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

h = np.sqrt(3) * 0.5

# 目盛線を描画
for i in range(0,10,1):
    ax.plot( [10*i, 10*i+(100-10*i)*0.5], [0, (100-10*i)*h], '--',color='grey', lw=0.5 )
    ax.plot( [10*i*0.5, 10*i], [10*i*h, 0], '--',color='grey', lw=0.5 )
    ax.plot( [10*i*0.5, 100-10*i*0.5], [10*i*h, 10*i*h], '--',color='grey', lw=0.5 )
# 外枠を描画
ax.plot( [0, 100],[0, 0], 'k-', lw=1 )
ax.plot( [0, 50],[0, 100 * h], 'k-', lw=1 )
ax.plot( [100, 50],[0, 100 * h], 'k-', lw=1 )
#軸ラベル
for i in range(2,10,2):
    ax.text( 0.5*(100-10*i)-6, h*10*(10-i), '%d0' % i, fontsize=12 )
    ax.text( 100-0.5*10*i+1, h*10*i,'%d0' % i, fontsize=12 )
    ax.text( 10*i-2.5, -5, '%d0' % i, fontsize=12 )
# 端成分
ax.text(45, 100 * h + 2, 'Wo', fontsize=16)
ax.text(-10, -2, 'En', fontsize=16)
ax.text(103, -2, 'Fs', fontsize=16)

for i in range(len(sheet)):
    df = pd.read_excel( 'EPMA analysis.xlsx', sheet_name=sheet[i] )
    # reject from total and cation
    data = df[ ( df['Cation'] > 3.95 ) & ( df['Cation'] < 4.05 ) & ( df['Total'] > 99.5 ) & ( df['Total'] < 104 ) ]

    Ca = data['CaO'] / 56.0774
    Mg = data['MgO'] / 40.3044
    Fe = data['FeO'] / 71.844 

    Wo = 100 * Ca / ( Ca + Fe + Mg )
    En = 100 * Mg / ( Ca + Fe + Mg )
    Fs = 100 * Fe / ( Ca + Fe + Mg )

    x = Wo * 0.5 + Fs
    y = Wo * h

    ax.plot( x, y, '.')

fig.savefig('cpx.png', dpi=600)
```
