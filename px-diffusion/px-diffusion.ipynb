```python
# -*- coding: utf-8 -*-
# import libraries
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

x = np.arange(0.01, 10.0, 0.01)

TempCmin = 900
TempCmax = 1100
TempK = np.arange(TempCmin+273, TempCmax+273, 50)

fO2 = 10**(-10)
R = 8.31
day = 60 * 60 * 24

fig = plt.figure(figsize=(4,3))
plt.subplots_adjust(hspace=0)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
# plot Mg# content
ax = fig.add_subplot(111)
#ax.set_title('Residence time', loc='center', fontsize=9)
c = ['r', 'b', 'g', 'k']
style = ['solid', 'dashed', 'dashdot', 'dotted']
for i in range(len(TempK)):
    D = 1.12 * ( 10**(-6) ) * ( fO2**0.053 ) * math.exp( -308000 / ( R * TempK[i] ) )
    t = ( 1.085 * ( ( x*10**(-6) )**2 ) / D ) / day
    ax.plot( x, t, linestyle='solid', color='k', linewidth=0.75, label=str(TempK[i]-273) )
#ax.set_xlim(55, 95)
#ax.set_ylim(0, 0.1)
#plt.gca().yaxis.set_major_formatter(plt.FormatStrFormatter('%.1f'))
ax.set_yscale('log')
ax.set_xlabel('Square diffused distance ($\mu$m)')
ax.set_ylabel('Residence time (days)')

#ax.legend(edgecolor='w', fancybox=False, fontsize=7)

# save figure
plt.savefig('residencetime.jpg', bbox_inches="tight", dpi=300)
```
